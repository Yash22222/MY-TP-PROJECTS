```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
```

```python
df = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data" # loading dataset
```

```python
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
df = pd.read_csv(df, names=column_names)
```

```python
df.groupby('species')['sepal_length'].mean().plot(kind='bar')
plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Box Plot of Sepal Length for Each Species')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='petal_width', data=df, palette='pastel')
plt.title('Violin Plot of Petal Width for Each Species')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True)
plt.title('Heatmap of Correlation Between Iris Features')
plt.show()
```

```python
px.scatter(df, x="sepal_length", y="sepal_width", color="species", title="Scatter Plot of Sepal Length and Sepal Width", labels={
    "sepal_length": "Sepal Length",
    "sepal_width": "Sepal Width",
    "species": "Species"
})
```

```python
px.scatter(df, x="sepal_length", y="petal_length", color="species", title="Scatter Plot of Sepal Length and Petal Length", labels={
    "sepal_length": "Sepal Length",
    "petal_length": "Petal Length",
    "species": "Species"
})
```

```python
plt.figure(figsize=(10, 6))
df['species'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Pie Chart of Iris Species')
plt.show()
```
